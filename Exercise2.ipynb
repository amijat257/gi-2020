```python
import pysam

merged_tumor = pysam.AlignmentFile("/sbgenomics/project-files/merged-tumor.bam")

num_of_reads = 0
num_of_unmapped_reads = 0
num_of_zero_mapping_quality = 0
average_mapping_quality = 0
average_mapping_quality_filtered = 0

for read in merged_tumor:
    num_of_reads += 1
    
    if read.is_unmapped:
        num_of_unmapped_reads += 1
    
    if read.mapping_quality == 0:
        num_of_zero_mapping_quality += 1
    else:
        average_mapping_quality_filtered += read.mapping_quality
        
    average_mapping_quality += read.mapping_quality

merged_tumor.close()

print("Total number of reads: " + str(num_of_reads))
print("Number of unmapped reads: " + str(num_of_unmapped_reads))
print("Number of reads with zero mapping quality: " + str(num_of_zero_mapping_quality))
print("Average mapping quality for all the reads: " + str(average_mapping_quality/num_of_reads))
print("Average mapping quality if reads with zero mapping quality are filtered out: " + str(average_mapping_quality_filtered/(num_of_reads-num_of_zero_mapping_quality)))
```
